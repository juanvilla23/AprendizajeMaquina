```python
import pandas as pd
accidentes = pd.read_csv("../Datos/Raw/accidentes.csv", parse_dates=["TW"])
```

```python
info_calidad = pd.DataFrame({
    'Tipo de Dato': accidentes.dtypes,
    'Valores Nulos': accidentes.isnull().sum(),
    'Porcentaje Nulos (%)': (accidentes.isnull().sum() / len(accidentes)) * 100
})

# Ordenar de mayor a menor porcentaje de nulos para ver qué columnas están más vacías
info_calidad = info_calidad.sort_values(by='Porcentaje Nulos (%)', ascending=False)
print(info_calidad)
```

```python
total_duplicados_id = accidentes.duplicated(subset=['TW', 'BARRIO']).sum()
print(f"Cantidad de duplicados: {total_duplicados_id}")
```

```python
# Lista de columnas categóricas que realmente están en el DataFrame
columnas_validas = [col for col in ['BARRIO', 'Dia_sem'] if col in accidentes.columns]

# 1. Ver el listado y frecuencias de los datos categóricos
for col in columnas_validas:
    print(f"\n--- Valores únicos y frecuencias en: {col} ---")
    print(accidentes[col].value_counts(dropna=False))

print("\n" + "="*50 + "\n")

# 2. Validar consistencia de horas (esta sí debe funcionar perfectamente)
horas_invalidas = accidentes[(accidentes['Hora'] < 0) | (accidentes['Hora'] > 23)]
print(f"Filas con horas fuera de rango (0-23): {len(horas_invalidas)}")
```

```python
import pandas as pd

accidentes = pd.read_csv("../Datos/Raw/accidentes.csv", parse_dates=["TW"])
clima = pd.read_csv("../Datos/Raw/clima.csv", parse_dates=["TW"])
raw = pd.read_csv("../Datos/Raw/raw_accidentes.csv", parse_dates=["TW"])
```

```python
import gc

filas_antes = len(clima)
clima.dropna(subset=["summary"], inplace=True)
gc.collect()

print(f"Filas eliminadas (sin summary): {filas_antes - len(clima):,}")
print(f"Filas restantes: {len(clima):,}")
```

```python
mediana = clima["windBearing"].median()
clima["windBearing"] = clima["windBearing"].fillna(mediana)

print(f"Mediana usada: {mediana}")
print(f"Nulos restantes: {clima['windBearing'].isnull().sum()}")
```

```python
import gc

filas_antes = len(clima)
clima.drop(clima.index[clima["BARRIO"] == "sinnombre"], inplace=True)
gc.collect()

print(f"Filas eliminadas (sinnombre): {filas_antes - len(clima):,}")
print(f"Filas restantes: {len(clima):,}")
```

```python
numericas = clima.select_dtypes(include="number")
negativos = {}

for col in numericas.columns:
    mask = numericas[col] < 0
    if mask.any():
        negativos[col] = mask.sum()

if negativos:
    print("Columnas con valores negativos:\n")
    for col, cantidad in negativos.items():
        pct = (cantidad / len(clima) * 100).round(4)
        min_val = clima[col].min()
        print(f"  {col}: {cantidad:,} registros ({pct}%) — mínimo: {min_val}")
else:
    print("No hay valores negativos en ninguna columna numérica.")

```

```python
mediana = clima["windBearing"].median()
nulos_antes = clima["windBearing"].isnull().sum()

clima["windBearing"] = clima["windBearing"].fillna(mediana)

print(f"Mediana windBearing: {mediana}")
print(f"Nulos rellenados: {nulos_antes:,}")
print(f"Nulos restantes: {clima['windBearing'].isnull().sum():,}")
```

```python
columnas_eliminar = ["icon", "apparentTemperature"]
columnas_a_borrar = [c for c in columnas_eliminar if c in clima.columns]

clima.drop(columns=columnas_a_borrar, inplace=True)

print(f"Columnas eliminadas: {columnas_a_borrar}")
print(f"Total columnas: {len(clima.columns)}")
print(clima.columns.tolist())
```

```python
nulos = clima.isnull().sum()
porcentaje = (clima.isnull().mean() * 100).round(4)
tabla_nulos = (
    pd.DataFrame({"Nulos": nulos, "Porcentaje (%)": porcentaje})
    .sort_values("Nulos", ascending=False)
)
print(f"Total filas: {len(clima):,}")
print(f"Filas con al menos un nulo: {clima.isnull().any(axis=1).sum():,}\n")
tabla_nulos
```

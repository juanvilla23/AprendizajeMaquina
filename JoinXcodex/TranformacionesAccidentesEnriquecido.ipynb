```python
import pandas as pd
from pathlib import Path

# Ruta al archivo accidentes.csv
RUTA_ACCIDENTES = Path("../Datos/Staging/accidentes_staging_enriquecido.csv")

# Leer el CSV
accidentes = pd.read_csv(RUTA_ACCIDENTES, parse_dates=["TW"])

# Mostrar primeras 5 filas
accidentes.head()
```

```python
print(accidentes.columns)
```

## Tabla `features_accidentes`

Copia del dataset con One-Hot de `CLASE`, sin `HORA` ni `Dia_sem` (ya representadas con `hora_sin`/`hora_cos` y `dia_semana_sin`/`dia_semana_cos`).

```python
COLUMNAS_ELIMINAR = ["HORA", "Dia_sem"]

faltantes = [c for c in COLUMNAS_ELIMINAR if c not in accidentes.columns]
if faltantes:
    print(f"Advertencia: no se encontraron columnas {faltantes}")

features_accidentes = accidentes.drop(columns=COLUMNAS_ELIMINAR, errors="ignore").copy()

print("Columnas eliminadas (información duplicada):")
for col in COLUMNAS_ELIMINAR:
    if col in accidentes.columns:
        print(f"  - {col}")

print(f"\naccidentes:          {accidentes.shape[0]:,} filas × {accidentes.shape[1]} columnas")
print(f"features_accidentes: {features_accidentes.shape[0]:,} filas × {features_accidentes.shape[1]} columnas")
print(f"\nColumnas de features_accidentes ({len(features_accidentes.columns)}):")
print(features_accidentes.columns.tolist())

display(features_accidentes.head())
```

## Eliminar `TIPO_GEOCOD`

Tiene miles de categorías casi únicas (geocodificación muy específica); no conviene para el modelo. Se quita de `features_accidentes`.

```python
COL_ELIMINAR = "TIPO_GEOCOD"

cols_antes = len(features_accidentes.columns)
filas = len(features_accidentes)

print("ANTES:")
print(f"  Columnas: {cols_antes}")
print(f"  ¿Existe {COL_ELIMINAR}? {COL_ELIMINAR in features_accidentes.columns}")

features_accidentes = features_accidentes.drop(columns=[COL_ELIMINAR], errors="ignore")

cols_despues = len(features_accidentes.columns)

print("\nDESPUÉS:")
print(f"  Columnas: {cols_despues}  (cambio: {cols_despues - cols_antes})")
print(f"  Filas: {filas:,} (sin cambio)")
print(f"\nColumnas actuales ({cols_despues}):")
print(features_accidentes.columns.tolist())

display(features_accidentes.head())
```

```python
COL_ELIMINAR = "OBJECTID"

cols_antes = len(features_accidentes.columns)

print("ANTES:")
print(f"  Columnas: {cols_antes}")

features_accidentes = features_accidentes.drop(columns=[COL_ELIMINAR], errors="ignore")

cols_despues = len(features_accidentes.columns)

print("\nDESPUÉS:")
print(f"  Columnas: {cols_despues}  (cambio: {cols_despues - cols_antes})")
print(f"  Filas: {len(features_accidentes):,}")
print(f"\nColumnas actuales ({cols_despues}):")
print(features_accidentes.columns.tolist())

display(features_accidentes.head())
```

## Eliminar `DIRECCION` y `DIRECCION_ENC`

Texto de dirección (original y codificada); redundante con `BARRIO`, `COMUNA` y variables temporales.

```python
COLUMNAS_ELIMINAR = ["DIRECCION", "DIRECCION_ENC"]

cols_antes = len(features_accidentes.columns)

print("ANTES:")
print(f"  Columnas: {cols_antes}")

features_accidentes = features_accidentes.drop(columns=COLUMNAS_ELIMINAR, errors="ignore")

cols_despues = len(features_accidentes.columns)

print("\nDESPUÉS:")
print(f"  Columnas: {cols_despues}  (cambio: {cols_despues - cols_antes})")
print(f"  Eliminadas: {COLUMNAS_ELIMINAR}")
print(f"  Filas: {len(features_accidentes):,}")
print(f"\nColumnas actuales ({cols_despues}):")
print(features_accidentes.columns.tolist())

display(features_accidentes.head())
```

## Eliminar `CLASE` y `GRAVEDAD`

Quitamos la columna original y, si existen, las dummies `clase_*` y `gravedad_*` (no se usan como predictores en el join con clima).

```python
cols_antes = len(features_accidentes.columns)

# Texto original + posibles columnas One-Hot
cols_clase_gravedad = [
    c for c in features_accidentes.columns
    if c in ("CLASE", "GRAVEDAD")
    or c.startswith("clase_")
    or c.startswith("gravedad_")
]

print("ANTES:")
print(f"  Columnas: {cols_antes}")
print(f"  A eliminar ({len(cols_clase_gravedad)}):")
for c in cols_clase_gravedad:
    print(f"    - {c}")

features_accidentes = features_accidentes.drop(columns=cols_clase_gravedad, errors="ignore")

cols_despues = len(features_accidentes.columns)

print("\nDESPUÉS:")
print(f"  Columnas: {cols_despues}  (cambio: {cols_despues - cols_antes})")
print(f"  Filas: {len(features_accidentes):,}")
print(f"\nColumnas actuales ({cols_despues}):")
print(features_accidentes.columns.tolist())

display(features_accidentes.head())
```

## Eliminar columnas de fecha/hora redundantes

Quitamos `FECHA`, `Mes`, `Dia_Sem_REAL` y `hora_dia` porque ya están representadas en `TW`, `mes_num`, `dia_semana` y las variables `hora_sin` / `hora_cos`.

```python
COLUMNAS_ELIMINAR = ["FECHA", "Mes", "Dia_Sem_REAL", "hora_dia", "PERIODO","comuna","Dia","Hora_num","mes_num","dia_semana"]

cols_antes = len(features_accidentes.columns)

print("ANTES:")
print(f"  Columnas: {cols_antes}")

features_accidentes = features_accidentes.drop(columns=COLUMNAS_ELIMINAR, errors="ignore")

cols_despues = len(features_accidentes.columns)

print("\nDESPUÉS:")
print(f"  Columnas: {cols_despues}  (cambio: {cols_despues - cols_antes})")
print(f"  Eliminadas: {COLUMNAS_ELIMINAR}")
print(f"  Filas: {len(features_accidentes):,}")
print(f"\nColumnas actuales ({cols_despues}):")
print(features_accidentes.columns.tolist())

display(features_accidentes.head())
```
